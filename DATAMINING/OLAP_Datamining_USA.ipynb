# **KIỂM TRA MISSING VALUE**

```python
df.columns
```

```python
def find_missing_percent(df):
  miss_df = pd.DataFrame({"ColumnName" : [], "TotalMissingVals" : [], "PercentMissing" : []})
  nrows = df.shape[0]
  cols = df.columns

  for col in cols:
    missing_count = df[col].isnull().sum()
    if(missing_count > 0):
      missing_info = {"ColumnName" : [col], "TotalMissingVals" : missing_count, "PercentMissing" : missing_count/nrows * 100}
      miss_df = pd.concat([miss_df, pd.DataFrame.from_dict(missing_info)], ignore_index= True)
  return miss_df

missing_df = find_missing_percent(df)
```

```python
display(missing_df.sort_values(by="PercentMissing", axis=0, ascending=False))
plt.figure(figsize=(16,9))
sns.heatmap(df.isnull())
```

# **DATA MINING**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Split the data into training and testing sets by 8:2
train, test = train_test_split(cleaned_df, test_size = 0.2, random_state = 42)
print(train.shape, ',', test.shape)
```

```python
targets = train['total']
features = train.drop('total', axis = 1)

x_train, x_valid, y_train, y_valid = train_test_split(features, targets, random_state=42)
print(x_train.shape, y_train.shape)
print(x_valid.shape, y_valid.shape)
```

# **DECISION TREE**

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
# Khởi tạo mô hình Decision Tree
dtc_model = DecisionTreeRegressor(random_state = 42)

# Huấn luyện mô hình trên tập huấn luyện chính
dtc_model.fit(x_train, y_train)

# Dự đoán trên tập xác thực
dtc_y_pred = dtc_model.predict(x_valid)

# Đánh giá mô hình
dtc_mse = mean_squared_error(y_valid, dtc_y_pred)
dtc_rmse = np.sqrt(dtc_mse)
dtc_mae = mean_absolute_error(y_valid, dtc_y_pred)
dtc_r_squared_train = dtc_model.score(x_train, y_train)
dtc_r_squared_valid = dtc_model.score(x_valid, y_valid)

# In kết quả
print("Mean Squared Error (MSE):", dtc_mse)
print("Root Mean Squared Error (RMSE):", dtc_rmse)
print("Mean Absolute Error (MAE):", dtc_mae)
print("R-squared - Train:", dtc_r_squared_train)
print("R-squared - Validation:", dtc_r_squared_valid)
```

```python
# Lấy thông tin của 10 đơn hàng tiếp theo để dự đoán giá trị
dtc_next_order = train.head(10)

# Lấy các đặc trưng cho 10 đơn hàng tiếp theo
dtc_next_order_features = dtc_next_order.drop('total', axis=1)

# Dự đoán giá trị của 10 đơn hàng tiếp theo
dtc_next_order_pred = dtc_model.predict(dtc_next_order_features)

# In ra giá trị dự đoán của 10 đơn hàng tiếp theo
print("Giá trị dự đoán tổng tiền (total) của 10 đơn hàng tiếp theo là:")
print(dtc_next_order_pred)
```

```python
# Scatter plot cho dự đoán trên tập xác thực
plt.scatter(y_valid, dtc_y_pred, color='blue', label='Validation Prediction')

# Scatter plot cho dự đoán của 10 đơn hàng tiếp theo
plt.scatter(x = dtc_next_order_features.index, y = dtc_next_order_pred,
            color='red', label='Next Orders Prediction')

plt.title('Decision Tree Regression')
plt.xlabel('Index')
plt.ylabel('Total')
plt.legend()
plt.show()
```

```python
# Tạo biểu đồ cây quyết định
plt.figure(figsize=(20, 10))  # Đặt kích thước hình ảnh
plot_tree(dtc_model, max_depth=4, feature_names=x_train.columns, filled=True, fontsize=10)
plt.show()
```
